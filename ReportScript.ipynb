### Cache Sorting

```python
import requests
import datetime
import pandas as pd
import numpy as np

#get all data from jsPsych repo
#note: this excludes anything in jsPsych -contrib repo

def fetch_api(url):
  final_data = []

  while url:
    response = requests.get(url)

    data = response.json()
    for item in data:
      final_data.append(item)

    #check for another response page
    if 'next' in response.links:
      url = response.links['next']['url']
    else:
      break

  return final_data
```

```python
# fix: sort pull_requests yearly as well; see if we can do this without having to comb through initially

def sort_entries(cache):
    issues = []
    issue_index = []

    prs = []
    pr_index = []

    year_index = []

    year_cache = [[[]],[[],[]]]
    #first sublist: issues
    #second sublist: prs
    #in each sublist -
      #first subsublist: index no.
      #second subsublist: entry

    for item in cache:
      if 'pull_request' in item:
        dest_list = 1
      else:
        dest_list = 0

      dt = datetime.datetime.strptime(item['created_at'], "%Y-%m-%dT%H:%M:%SZ")
      year = dt.year
      url = item['url']
      num = int(url.split('/')[-1])

      #fix: dear god there has to be a more elegant way to do this conditionals

      if item==cache[-1]:

        #set-up for last update
        year_cache[dest_list][0].append(num)
        year_cache[dest_list][1].append(item)

      if item==cache[-1] or (item!=cache[0] and year not in year_index):

        #update sorted outputs
        issue_index.append(year_cache[0][0])
        issues.append(year_cache[0][1])
        pr_index.append(year_cache[1][0])
        prs.append(year_cache[1][1])

      if year not in year_index:

        #add new year
        year_index.append(year)

        #reset year_cache
        year_cache[dest_list] = [[num],[item]]
        year_cache[abs(dest_list-1)] = [[],[]]

      else:
        year_cache[dest_list][0].append(num)
        year_cache[dest_list][1].append(item)

    return [year_index, issue_index, issues, pr_index, prs]
```

```python
def comments_by_issue(comments):
    comments_sorted = []
    comments_issnums = []

    for com in comments:
      iss_url = com['issue_url']
      iss_num = int(iss_url.split('/')[-1])
      if iss_num in comments_issnums:
        comments_sorted[comments_issnums.index(iss_num)].append(com)
      else:
        comments_sorted.append([com])
        comments_issnums.append(iss_num)

    return [comments_issnums, comments_sorted]
```

### Dataframe Set-ups


```python
#just defining functions for last cell; literally run it whenever you want, i do not care
#you could even do it for fun

def seconds(tdelta):
  return tdelta.total_seconds()

total_seconds = np.vectorize(seconds)

def wait_breakdown(sec_diff):

  days = sec_diff / 3600 / 24
  day_diff = int(days)
  hour_diff = int((days - day_diff) * 24)
  min_diff = int(((days - day_diff) * 24 - hour_diff) * 60)

  return f'{day_diff} days, {hour_diff} hours, and {min_diff} minutes'
```

```python
#only works if all the above is run

def report_frame(issues, issue_index, commented_issues, comments):

  report = []

  for y, year in enumerate(issues):
    data = []

    for i, iss in enumerate(year):
      iss_dic = {}
      iss_dic['Issue No.'] = issue_index[y][i]

      #get date created
      created_at = datetime.datetime.strptime(iss['created_at'], "%Y-%m-%dT%H:%M:%SZ")
      iss_dic['Created At'] = created_at
      dates = np.array([created_at])

      #get comment dates if there are comments
      #NOTE: commented_issues is used here to connect comments to their issues according to issue number
      #   - this helps get comment dates and calculate average interaction time later in this same cell
      if iss_dic['Issue No.'] in commented_issues:
        iss_comments = comments[commented_issues.index(iss_dic['Issue No.'])]
        com_qt = len(iss_comments)
        iss_dic['Number of Comments'] = com_qt
        for com in iss_comments:
          com_date = datetime.datetime.strptime(com['created_at'], "%Y-%m-%dT%H:%M:%SZ")
          dates = np.append(dates, com_date)
      else:
        iss_dic['Number of Comments'] = 0

      #get date closed if closed
      if iss['closed_at']:
        closed_at = datetime.datetime.strptime(iss['closed_at'], "%Y-%m-%dT%H:%M:%SZ")
        iss_dic['Closed At'] = closed_at
        dates = np.append(dates, closed_at)
      else:
        iss_dic['Closed At'] = 'Still Open'

      #get average interaction time on issue
      #FIX: check for redundancy
      if len(dates) > 1:
        time_diffs = dates[1:] - dates[:-1]
        time_diffs = total_seconds(time_diffs)
        avg_time_diff = np.mean(time_diffs)
        iss_dic['Average Interaction Time'] = wait_breakdown(avg_time_diff)

        if iss_dic['Closed At'] == 'Still Open':
          today = datetime.datetime.today()
          open_time = seconds(today - created_at)
          iss_dic['Ignored For'] = wait_breakdown(open_time)
        else:
          iss_dic['Ignored For'] = 'Closed'

      else:
        today = datetime.datetime.today()
        ignore_time = seconds(today - created_at)
        iss_dic['Ignored For'] = wait_breakdown(ignore_time)
        iss_dic['Average Interaction Time'] = 'No interaction'

      data.append(iss_dic)

    report.append(pd.DataFrame(data))

  return report
```

```python
# [X] visualize issue opening per year / per every 3 months

# [Samip] goes through the issue report and count how many in each quarter for each year and return a list where each item is a year, and inside each year, we have counts for each of the 4 quarters.

def issues_by_quarter(issue_report):
    quarters = [1, 4, 7, 10]
    issues_per_quarter = []

    for year_data in issue_report:
        year_data = year_data.to_dict('records')
        year_quarter_counts = [0, 0, 0, 0]

        for issue in year_data:
            created_month = issue['Created At'].month
            quart = max(month for month in quarters if month <= created_month)
            quarter_index = quarters.index(quart)
            year_quarter_counts[quarter_index] += 1

        issues_per_quarter.append(year_quarter_counts)

    return issues_per_quarter
```

```python
import re

def avg_time_quarterly(issue_report):

  data = []
  quarters = [1, 4, 7, 10]

  for year in issue_report:
    data_year = [[], [], [], []]

    #turn Dataframe back into dictionary
    year = year.to_dict('records')

    for iss in year:
      if iss['Average Interaction Time'] != 'No interaction':
        quart = max(month for month in quarters if month <= iss['Created At'].month)

        #turns string back into aggregate minutes
        #some seconds get shaved out of consideration; can't be fix w.o changing issue_timedata() itself
        minutes = re.findall(r'-?\d+', iss['Average Interaction Time'])
        minutes = [int(num) for num in minutes]
        minutes = (minutes[0]*60*24) + (minutes[1]*60) + minutes[2]

        data_year[quarters.index(quart)].append(minutes)

    data.append([np.mean(quarter) for quarter in data_year])

  return data
```

```python
#characterization: issues with more commentary than usual
def gthan_ten(issue_report):
  gthan_list = []
  # entries = 0

  for year in issue_report:
    year = year.to_dict('records')
    for entry in year:
      if entry['Number of Comments'] > 10:
        gthan_list.append(entry)
        # print(entry['Issue No.'], entry['Created At'], entry['Number of Comments'], entry['Closed At'])
        # e = e + 1

  df = pd.DataFrame(gthan_list)
  print(f"There are {len(gthan_list)} issues with more than 10 comments")
  return df
```

```python
#characterization: issues with no commentary
def zero_comments(issue_report):
  zero_comments_list = []
  for year in issue_report:
    year = year.to_dict('records')
    for entry in year:
      if entry['Number of Comments'] == 0:
        zero_comments_list.append(entry)

  df = pd.DataFrame(zero_comments_list)
  print(f"There are {len(zero_comments_list)} issues with 0 comments")
  return df
```
